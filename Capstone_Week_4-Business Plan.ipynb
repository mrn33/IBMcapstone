## Mapping The 100 Best Coffee Shops For the Top 5 Cities for Coffee in the US

### Objective

### Many coffee enthusiats travel to unfamiliar regions for the opportunity to taste and experience the best coffee that a city can deliver.  According to CNBC.com the top 5 coffee cities in the US are the following: 

###### New York, NY, Seattle, WA, San Francisco, CA, Portland, OR, and Los Angeles, CA

## Pulling location data from Foursquare's api I would like to map out the top 100 coffee locations of five the aformentioned cities so that coffee enthusiats can plan their trips with these maps in mind.  For example rather than staying at a hotel just because it has a good rating, coffee lovers can utilize this application to make their travel plans so that they are in the areas most densly populated with high rated coffee shops, thus making the most of their trips!
